```python
# 列表推导式

counts = [j for i in df_liepin_gz['job.labels'].apply(lambda x:eval(x)).tolist() for j in i]
counts
```

```python

counts.count("制造业")
```

```python
coun = set(counts)
```

```python
[(i,counts.count(i)) for i in coun]
```

```python
import pyecharts.options as opts
from pyecharts.charts import WordCloud

data = [(i,counts.count(i)) for i in coun]


(
    WordCloud()
    .add(series_name="热点分析", data_pair=data, word_size_range=[6, 66])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="热点分析", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
    .render("hha.html")
)
```
